## Metadata

```python
obs = adata.obs.copy()
obs['cluster'] = kdata.obs['KD_label_combined'].values
obs['KD_label'] = 'NA'

for sample in haowen_label['sample'].unique():
    obs.loc[obs['sample'] == sample, 'KD_label'] =  haowen_label.loc[haowen_label['sample'] == sample, 'KD_label_new'].values
    
obs = obs.loc[:, ['orig_index', 'condition', 'phase_ref', 'cluster', 'KD_label']]
obs.columns = ['orig_index', 'condition', 'cell_cycle', 'cluster', 'KD_label']
obs
```

```python
obs_col = [
    ## required
    ('NAME', 'TYPE'), ##
    ('biosample_id', 'group'), ## 
    ('donor_id', 'group'), ##
    ('species', 'group'), ##
    ('species__ontology_label', 'group'), ##
    ('sex', 'group'), ##
    ('cell_type', 'group'),
    ('cell_type__ontology_label', 'group'),
    ('disease', 'group'), ##
    ('disease__ontology_label', 'group'), ##
    ('organ', 'group'), ## 
    ('organ__ontology_label', 'group'), ##
    ('library_preparation_protocol', 'group'), ##
    ('library_preparation_protocol__ontology_label', 'group'), ##

    ('orig_index', 'numeric'),
    ('condition', 'group'),
    ('cell_cycle', 'group'),
    ('cluster', 'group'),
    ('KD_label', 'group'),]

obs_df = pd.DataFrame('NA', index=expr_col, columns=pd.MultiIndex.from_tuples(obs_col))
obs_df
```

```python
# add value
obs_df['NAME'] = obs_df.index
obs_df['biosample_id'] = adata.obs['sample']
obs_df['donor_id'] = adata.obs['sample']
obs_df['species'] = 'NCBITaxon_9606'
obs_df['species__ontology_label'] = 'Homo Sapiens'
obs_df['sex'] = 'female'

obs_df['cell_type'] = 'CL_0000066'
obs_df['cell_type__ontology_label'] = 'epithelial cell'

obs_df['disease'] = 'MONDO_0004970'
obs_df['disease__ontology_label'] = 'adenocarcinoma'
obs_df['organ'] = 'UBERON_0000002'
obs_df['organ__ontology_label'] = 'uterine cervix'

obs_df['library_preparation_protocol'] = 'EFO_0030000'
obs_df['library_preparation_protocol__ontology_label'] = 'STARmap'

obs_df['orig_index'] = obs['orig_index']
obs_df['condition'] = obs['condition']
obs_df['cell_cycle'] = obs['cell_cycle']
obs_df['cluster'] = obs['cluster']
obs_df['KD_label'] = obs['KD_label']
obs_df
```

```python
obs_df.to_csv(os.path.join(out_path, 'metadata.csv'), index=False)
```
