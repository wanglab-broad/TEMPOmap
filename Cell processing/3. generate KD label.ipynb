## KD clustering (Combined)

```python
# Subset 
kd_genes = ['METTL14', 'METTL3', 'YTHDC1', 'YTHDC2', 'YTHDF1', 'YTHDF2', 'YTHDF3'] # ''YTHDC2','
kddata = adata[:, kd_genes]
kddata
```

```python
(kddata.X.sum(axis=1) == 0).sum()
```

```python
# Normalization scaling
sc.pp.normalize_total(kddata)
sc.pp.log1p(kddata)
kddata.raw = kddata

# Scale data to unit variance and zero mean
sc.pp.scale(kddata)
kddata.layers['scaled'] = kddata.X
```

### Dimensionality reduction

```python
# Run PCA
sc.tl.pca(kddata, svd_solver='arpack')

# Plot explained variance 
sc.pl.pca_variance_ratio(kddata, log=False)
```

```python
# Plot PCA
for i, gene in enumerate(kddata.var.index):
    print(gene)
    sc.pl.pca(kddata, color=gene)
```

```python
%%time
# Computing the neighborhood graph
n_neighbors = 10
n_pcs = 6

# sc.pp.neighbors(kddata, n_neighbors=n_neighbors, n_pcs=n_pcs, use_rep='X')
sc.pp.neighbors(kddata, n_neighbors=n_neighbors, n_pcs=n_pcs)

# Run UMAP
sc.tl.umap(kddata)
sc.pl.umap(kddata, color='sample')
```

```python
sc.pl.umap(kddata, color='total_counts')
```

```python
for gene in kddata.var.index:
    print(gene)
    sc.pl.umap(kddata, color=gene)
```

### Kmeans

```python
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

# Use PCA embedding
# extract pca coordinates
X_pca = kddata.obsm['X_pca']
# X_pca = kddata.X

# kmeans with k=8
kmeans = KMeans(n_clusters=2, random_state=1).fit(X_pca) 
kddata.obs['kmeans2'] = kmeans.labels_.astype(str)

# sc.pl.pca(kddata, color=['kmeans8'])
# sc.tl.dendrogram(adata, 'kmeans8')
sc.pl.heatmap(kddata, kddata.var.index, groupby='kmeans2', dendrogram=False, use_raw=False, cmap='bwr', vmin=-3, vmax=3)
```

```python
kddata.obs['KD_label_combined'] = kddata.obs['kmeans2']
kddata.obs['KD_label_combined'] = kddata.obs['KD_label_combined'].astype(object)

kddata.obs['KD_label_combined'] = kddata.obs['KD_label_combined'].map({'0': 'siYTHDC1',
                                                             '1': 'siControl'})
kddata.obs['KD_label_combined'] = kddata.obs['KD_label_combined'].astype('category')
sc.pl.heatmap(kddata, kddata.var.index, groupby='KD_label_combined', dendrogram=False, use_raw=False, cmap='bwr', vmin=-4, vmax=4)
```

```python
sc.pl.matrixplot(kddata, kddata.var.index, groupby='KD_label_combined', dendrogram=False, use_raw=False, cmap='bwr', vmin=-4, vmax=4)
```

```python
sc.pl.pca(kddata, color='KD_label_combined')
sc.pl.umap(kddata, color='KD_label_combined')
```

```python
adata.obs['KD_label_combined'] = kddata.obs['KD_label_combined']
```

### Kmeans

```python
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

# Use PCA embedding
# extract pca coordinates
X_pca = kddata.obsm['X_pca']

# kmeans with k=8
k = 10
kmeans = KMeans(n_clusters=k, random_state=5).fit(X_pca)
kddata.obs[f'kmeans{k}'] = kmeans.labels_.astype(str)

sc.pl.pca(kddata, color=[f'kmeans{k}'])
sc.pl.heatmap(kddata, kddata.var.index, groupby=f'kmeans{k}', dendrogram=False, use_raw=False, cmap='bwr', vmin=-4, vmax=4)
sc.pl.matrixplot(kddata, kddata.var.index, groupby=f'kmeans{k}', dendrogram=False, use_raw=False, cmap='bwr', vmin=-4, vmax=4)
```

```python
# merged 
kddata.obs['kmeans8'] = kddata.obs['kmeans8'].map({'0': '1',
                                                     '1': '2',
                                                     '2': '3', 
                                                     '3': '6', 
                                                     '4': '0', 
                                                     '5': '7', 
                                                     '6': '4', 
                                                     '7': '5'})
sc.pl.heatmap(kddata, kddata.var.index, groupby='kmeans8', dendrogram=False, use_raw=False, cmap='bwr', vmin=-4, vmax=4)
sc.pl.matrixplot(kddata, kddata.var.index, groupby='kmeans8', dendrogram=False, use_raw=False, cmap='bwr', vmin=-4, vmax=4)
```

```python
kddata.obs['KD_label'] = kddata.obs['kmeans8']
kddata.obs['KD_label'] = kddata.obs['KD_label'].astype(object)

kddata.obs['KD_label'] = kddata.obs['KD_label'].map({'0': 'siControl', 
                                                             '1': 'siMETTL14', 
                                                             '2': 'siMETTL3', 
                                                             '3': 'siYTHDC1', 
                                                             '4': 'siYTHDC2', 
                                                             '5': 'siYTHDF1', 
                                                             '6': 'siYTHDF2', 
                                                             '7': 'siYTHDF3'})
kddata.obs['KD_label'] = kddata.obs['KD_label'].astype('category')
sc.pl.heatmap(kddata, kddata.var.index, groupby='KD_label', dendrogram=False, use_raw=False, cmap='bwr', vmin=-4, vmax=4)
sc.pl.matrixplot(kddata, kddata.var.index, groupby='KD_label', dendrogram=False, use_raw=False, cmap='bwr', vmin=-4, vmax=4)
```

```python
sc.pl.pca(kddata, color='KD_label')
sc.pl.umap(kddata, color='KD_label')
```
